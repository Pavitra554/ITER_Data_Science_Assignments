4. Write a function to compute the component-wise mean of a list of vectors. Assert the condition that
   the vectors must be of same length.


```python
def component_mean(vectors):
    m = len(vectors)
    n = len(vectors[0])
    out = []
    for i in range(m):
        sum = 0
        for j in range(n):
            sum += vectors[j][i]
        out.append(sum/n)
    return out


vectors = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
component_mean(vectors)
```

6. Write a program to find median of a given list of integers. Combine both odd and even number of
   terms.


```python
def median(data):
    n = len(data)
    temp = sorted(data)

    if n % 2 != 0:
        return temp[n // 2]
    else:
        return (temp[n // 2-1]+temp[n // 2])/2


median([2, 54, 5, 4, 2, 6, 81, 2])
```

7. We have defined the function normal cdf. Write a program to invert normal cdf to find the value
   corresponding to a specified probability.


```python
def normal_cdf(x, mu = 0, sigma = 1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2
def inverse_normal_cdf(p,mu = 0,sigma = 1,tolerance = 0.00001):

    if mu != 0 or sigma != 1:
        return mu + sigma * inverse_normal_cdf(p, tolerance=tolerance)
    low_z = -10.0  # normal_cdf(-10) is (very close to) 0
    hi_z = 10.0  # normal_cdf(10) is (very close to) 1
    while hi_z - low_z > tolerance:
        mid_z = (low_z + hi_z) / 2  
        mid_p = normal_cdf(mid_z)
        if mid_p < p:
            low_z = mid_z  
        else:
            hi_z = mid_z  
    return mid_z
inverse_normal_cdf(0.8)
```

8. Plot the Normal PDFs using various value of µ and σ as mentioned below:
mu sigma
0   1
0   2
0   0.5 
-1  1
Use different line styles for each plot and compare the graphs thus obtained. You can use any range
for x-axis.


```python
import math
import matplotlib.pyplot as plt
def normal_pdf(x, mu=0, sigma=1):
    # Normal PDF f(x) = (1/√(2πσ2)) (e[-(x-μ)^2]/2σ^2)
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x - mu) ** 2 / (2 * sigma ** 2)) / (sqrt_two_pi*sigma))

xs = [x for x in range(-10, 10)]
plt.plot(xs, [normal_pdf(x, 0, 1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_pdf(x, 0, 2) for x in xs], '.', label='mu=0, sigma=2')
plt.plot(xs, [normal_pdf(x, 0, 0.5) for x in xs], '-.', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_pdf(x, -1, 1) for x in xs], ':', label='mu=-1, sigma=1')
plt.legend()
plt.show()
```
