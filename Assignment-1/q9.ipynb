```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10,200)
data1 = x*np.sin(x)
data2 = x**2*np.sin(x)
data3 = x**3*np.sin(x)
data4 = x**4*np.sin(x)

print(data1)
```

```python
plt.plot(x,data1, label='xsin(x)')
plt.plot(x,data2, label='x^2sin(x)')
plt.plot(x,data3, label='x^3sin(x)')
plt.plot(x,data4, label='x^4sin(x)')

plt.title('sine functions plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show();
```
