```python
# Pavitra Behara - 2041002041
# Question 1 and 2
salaries_and_tenures = [(83000, 8.7), (88000, 8.1), (48000, 0.7), (76000, 6), (
    69000, 6.5), (76000, 7.5), (60000, 2.5), (83000, 10), (48000, 1.9), (63000, 4.2)]

# Defining the groups
g1 = []  # Less Than Two
g2 = []  # Between two and five
g3 = []  # More than five

# Filtering elements in groups
for s, t in salaries_and_tenures:
    if t < 2:
        print(f'({s},{t}) --> Less Than Two')
        g1.append(s)
    elif t >= 2 and t <= 5:
        print(f'({s},{t}) --> Between Two and five')
        g2.append(s)
    else:
        print(f'({s},{t}) --> More sthan five')
        g3.append(s)

print("Average salary for less than two years of tenure:",
      sum(g1)/len(g1))
print("Average salary for two to five years of tenure:",
      sum(g2)/len(g2))
print("Average salary for more than five years of tenure:",
      sum(g3)/len(g3))
```

```python
import matplotlib.pyplot as plt
sal=[s[0] for s in salaries_and_tenures]
ten=[s[1] for s in salaries_and_tenures]
 # taking as x-axis
 # taking as y-axis
plt.scatter(ten,sal)
plt.xlabel("experience(yrs)")
plt.ylabel("slary($)")
plt.title("Salary vs Years of Experience")
plt.grid()  

# Correspondance:- Users with very less yrs of experience need more salary to learn probably
# users with more years of experience need more salary to get more resources or tools for
# established business purposes(assumption again) also more experience in Data Science Field
# might show more professionalism(Hence more salary)
# Avreage years of experience Dont show much rapid change as they might be in between jobs and
# might not be willing to commit to a single company. 
```
