```python
import matplotlib.pyplot as plt

#data
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions)
plt.xlabel('Year')
plt.ylabel('Mentions')
plt.title('Mentions per year')
plt.show()
```
