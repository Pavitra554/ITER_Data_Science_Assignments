```python
import matplotlib.pyplot as plt

s = input('Enter string : ').lower()

data = {}
for c in s:
    if c.isalpha():
        if c in data:
            data[c] = data.get(c)+1
        else:
            data[c] = 1
print(data)
```

```python
plt.bar(data.keys(),data.values())
plt.xlabel('Characters')
plt.ylabel('frequency')
plt.title('Frequency of characters in string')
plt.show()
```
