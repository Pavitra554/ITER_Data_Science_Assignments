```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([2,3,1])
b = np.array([1,3,2]) 

plt.scatter(a,b)
```

```python
m = 0 # slope
c = 0 # intercept
lr = 0.01 # learning rate
loop = 1000 # iterations


for e in range(loop):
    y = m*a+c                  # y_predicted = slope_current * a + intrecept_current

    s1 = 0
    s2 = 0
    for i in range(3):
        s1 += (a[i]*(b[i] - m*a[i] - c))
        s1 += (b[i] - m*a[i] - c)

        
    sg = -(2/3)*s1  # slope_gradient = (2 / 3) summition  i->1 to 3 (xi * (ai-mxi-c))
    ic = -(2/3)*s2  # intercept_gradient = - (2 / 3) summition  i->1 to 3 (xi * (ai-mxi-c))
    m=m-lr*sg  # slope_current = slope_current - learning_rate * slope_gradient
    c=c-lr*ic  # intercept_gradient = intercept_gradient - learning_rate * intercept_gradient
    # plt.plot(a,y)
plt.scatter(a,b)
plt.plot(a,y)
```
