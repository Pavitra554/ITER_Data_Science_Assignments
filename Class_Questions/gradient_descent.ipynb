```python
# # patial derivative

# # ((f(x+h),y)-f(x,y))/h
l=[2,3]
def f(l):
    return ((l[0]**2)*l[1]+2*l[1])

            
def partial_derivative(f,l,i,h=0.0001):
    w=[v_i+(h if j==i else 0) for j,v_i in enumerate(l)]
    return (f(w)-f(l))/h
            
def gradient(f,l,h=0.0001):
    for i in range(len(l)):
        yield partial_derivative(f,l,i)
            
new_list=[]
for i in gradient(f,l):
    new_list.append(i)
new_list
```
