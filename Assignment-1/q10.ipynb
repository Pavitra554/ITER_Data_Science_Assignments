```python
import matplotlib.pyplot as plt

#data

male_age = [53,51,71,31,33,39,52,27,54,30,64,26,21,54,52,20,59,32]
female_age = [53,65,68,21,75,46,24,63,61,24,49,41,39,40,25,54,42,32,48,23,23]
```

```python
fig,f = plt.subplots(1,2,figsize=(15,7))

f[0].hist(male_age)
f[0].set_title('Male Age')
f[0].set_xlabel('Age')
f[0].set_ylabel('Frequency')

f[1].hist(female_age)
f[1].set_title('Female Age')
f[1].set_xlabel('Age')
f[1].set_ylabel('Frequency')
plt.show()
```
